## hermirian to QAM-16 original signal

```python
import numpy as np

def reverse_hermitian_symmetry_matrix(hermitian_symmetric_matrix):
    """Recover the QAM symbols from a matrix of Hermitian-symmetric OFDM symbols.
    
    Each row of the input matrix represents a 16-subcarrier OFDM symbol.
    """
    num_rows = hermitian_symmetric_matrix.shape[0]  # Get the number of rows
    qam_symbols_matrix = np.zeros((num_rows, 3), dtype=complex)  # Initialize the output array
    
    for i in range(num_rows):
        # Extract the original QAM symbols from each row
        qam_symbols_matrix[i, :] = hermitian_symmetric_matrix[i, 1:4]
    
    return qam_symbols_matrix

qam_symbols_matrix = reverse_hermitian_symmetry_matrix(fft_ofdm_symbols)

# # Print the recovered QAM symbols for each row
# print(qam_symbols_matrix)

# Concatenate all rows into a single row
QAM_16 = np.concatenate(qam_symbols_matrix)

# Print the concatenated row
#print(QAM_16)
```

```python
def qam4_to_binary(symbols):
    """Convert QAM-16 symbols back to a binary list, replace unmatched symbols with '1010'."""
    binary_list = []
    for RI in symbols:
        R = np.real(RI)
        I = np.imag(RI)

        if R >=0 and I>=0 :
            binary_list.append('00')
        elif R < 0 and I>=0 :
            binary_list.append('10')
        elif R < 0 and I < 0 :
            binary_list.append('11')
        elif R >=0 and I< 0 :
            binary_list.append('01')
        
    # Join the list of binary strings and convert them into individual bits
    return binary_list

binary_output = qam4_to_binary(QAM_16)
# print("Binary Output:", binary_output)
# Convert list to a single string
binary_string = ''.join(binary_output)

print(binary_string)
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# Create the figure and axes
fig, ax = plt.subplots(figsize=(10, 8))

# Draw encoder blocks
for i in range(4):
    # Left side rectangles (Encoder)
    ax.add_patch(patches.Rectangle((i * 1.5, 8 - i * 2), 1, 2, edgecolor='blue', facecolor='lightblue', lw=2))
    # Downward arrow for pooling
    if i < 3:
        ax.arrow(i * 1.5 + 0.5, 8 - i * 2 - 0.2, 0, -1.5, head_width=0.2, head_length=0.3, fc='black', ec='black')

# Draw bottleneck block
ax.add_patch(patches.Rectangle((4.5, 2), 1, 2, edgecolor='purple', facecolor='mediumpurple', lw=2))
ax.text(5, 2.8, "Bottleneck", ha='center', va='center', fontsize=10, color='white')

# Draw decoder blocks
for i in range(3):
    # Right side rectangles (Decoder)
    ax.add_patch(patches.Rectangle((6 + i * 1.5, 2 + i * 2), 1, 2, edgecolor='green', facecolor='lightgreen', lw=2))
    # Upward arrow for upsampling
    ax.arrow(6 + i * 1.5 + 0.5, 2 + i * 2 - 0.2, 0, -1.5, head_width=0.2, head_length=0.3, fc='black', ec='black')
    # Skip connection arrows
    ax.plot([1.5 * i + 1, 6 + i * 1.5], [8 - i * 2 + 1, 4 + i * 2], linestyle='--', color='gray')

# Add input and output labels
ax.text(0, 9, "Input", fontsize=12, color='black', ha='center')
ax.text(9.5, 7, "Output", fontsize=12, color='black', ha='center')

# Connect input and output
ax.arrow(-0.5, 8.5, 0.5, -0.5, head_width=0.2, head_length=0.3, fc='black', ec='black')
ax.arrow(8.5, 6.5, 1, 0.5, head_width=0.2, head_length=0.3, fc='black', ec='black')

# Add labels for layers
ax.text(1, 8.5, "Encoder Block 1", fontsize=10, ha='center', color='blue')
ax.text(2.5, 6.5, "Encoder Block 2", fontsize=10, ha='center', color='blue')
ax.text(4, 4.5, "Encoder Block 3", fontsize=10, ha='center', color='blue')
ax.text(7, 2.5, "Decoder Block 1", fontsize=10, ha='center', color='green')
ax.text(8.5, 4.5, "Decoder Block 2", fontsize=10, ha='center', color='green')
ax.text(10, 6.5, "Decoder Block 3", fontsize=10, ha='center', color='green')

# Set axis limits and hide them
ax.set_xlim(-1, 11)
ax.set_ylim(0, 10)
ax.axis('off')

# Show the block diagram
plt.show()
```
